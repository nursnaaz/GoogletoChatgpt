```python
def preprocess(text):
    text = re.sub('From:.*\n?', '',text )
    text = re.sub('Lines:.*\n?', '',text )
    text = re.sub('Keywords:.*\n?', '',text )
    text = re.sub('Article-I.D.:.*\n?', '',text )
    text = re.sub('Distribution:.*\n?', '',text )
    text = re.sub('Reply-To:.*\n?', '',text )
    text = re.sub('Subject: ', '',text )
    text = re.sub(r'--\n.*', '', text,re.DOTALL)
    return text
```

```python
import re
data_preprocess = []
for news in data:
    text = preprocess(news)
    text = preprocess_text(text)
    data_preprocess.append(text)
    

    

```
